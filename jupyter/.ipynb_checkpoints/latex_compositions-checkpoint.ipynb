```python
import os
import glob
import numpy as np
import pandas as pd
```

```python
src = '../manuscript2/ect/'
ects_file = sorted(glob.glob(src + '*.pdf'))
seed_file_X = sorted(glob.glob(src + '*X*.png'))
seed_file_Y = sorted(glob.glob(src + '*Y*.png'))
seed_file_Z = sorted(glob.glob(src + '*Z*.png'))
T = len(seed_file_X)-1
```

```python
header = "\\documentclass[tikz]{standalone}\n"
header += "\\begin{document}\n"
header += "\\begin{tikzpicture}[font=\\sffamily]\n"

footer = "\end{tikzpicture}\n"
footer += "\\end{document}"
```

```python
48%T
```

```python
i = 96

seed_X = "\\node[anchor=south west,inner sep=10] (seedX) at (0,0) {\\includegraphics[width=0.7\\textwidth]{"
seed_X += os.path.split(seed_file_X[np.min([i,len(seed_file_X)-1])])[1]
seed_X += "}};\n"

seed_Y = "\\node[anchor=west,inner sep=10] (seedY) at (seedX.east) {\\includegraphics[width=0.7\\textwidth]{"
if i <= T:
    j = 0
elif i > T and i < 2*T:
    j = i%T
else:
    j = -1
seed_Y += os.path.split(seed_file_Y[j])[1]
seed_Y += "}};\n"

seed_Z = "\\node[anchor=west,inner sep=10] (seedZ) at (seedY.east) {\\includegraphics[width=0.7\\textwidth]{"
if i <= 2*T:
    j = 0
elif i == 3*T:
    j = -1
else:
    j = i%T
seed_Z += os.path.split(seed_file_Z[j])[1]
seed_Z += "}};\n"
seed = seed_X + seed_Y + seed_Z
```

```python
ect = "\\node[anchor=north, inner ysep=10] (ect) at (seedY.south){\\includegraphics[width=3\\textwidth]{"
ect += os.path.split(ects_file[i])[1]
ect += "}};\n"
body = seed + ect
```

```python
filename = src + "ect_{:03d}.tex".format(i)
with open(filename, "w") as tex:
    tex.write(header + body + footer)
```

```python
for i in range(len(ects_file)):
    seed_X = "\\node[anchor=south west,inner sep=10] (seedX) at (0,0) {\\includegraphics[width=0.7\\textwidth]{"
    seed_X += os.path.split(seed_file_X[np.min([i,len(seed_file_X)-1])])[1]
    seed_X += "}};\n"

    seed_Y = "\\node[anchor=west,inner sep=10] (seedY) at (seedX.east) {\\includegraphics[width=0.7\\textwidth]{"
    if i <= T:
        j = 0
    elif i > T and i < 2*T:
        j = i%T
    else:
        j = -1
    seed_Y += os.path.split(seed_file_Y[j])[1]
    seed_Y += "}};\n"

    seed_Z = "\\node[anchor=west,inner sep=10] (seedZ) at (seedY.east) {\\includegraphics[width=0.7\\textwidth]{"
    if i <= 2*T:
        j = 0
    else:
        j = i%T
    seed_Z += os.path.split(seed_file_Z[j])[1]
    seed_Z += "}};\n"
    
    ect = "\\node[anchor=north, inner ysep=10] (ect) at (seedY.south){\\includegraphics[width=3\\textwidth]{"
    ect += os.path.split(ects_file[i])[1]
    ect += "}};\n"
    
    body = seed_X + seed_Y + seed_Z + ect
    
    filename = src + "ect_{:03d}.tex".format(i)
    with open(filename, "w") as tex:
        tex.write(header + body + footer)
```
